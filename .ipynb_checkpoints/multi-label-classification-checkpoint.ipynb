```python
import matplotlib
matplotlib.use("TkAgg")
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import random
import pickle
import cv2
import os
from imutils import paths

from Model import SmallerVGGNet

from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from keras.preprocessing.image import img_to_array
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from keras.models import load_model
```

```python
EPOCHS = 75
LEARNING_RATE = 0.001
BATCH_SIZE = 32
IMAGE_DIMS = (96, 96, 3)
```

```python
paths = sorted(list(paths.list_images("dataset")))
```

```python
random.seed(2)
random.shuffle(paths)
```

```python
paths[0]
```

```python
#data, labels = ([] for _ in range(2))
data = []
labels = [] 

for path in paths:
    img = cv2.imread(path)
    img = cv2.resize(img, (IMAGE_DIMS[1], IMAGE_DIMS[0]))
    img = img_to_array(img)
    data.append(img)
    
    label = path.split(os.path.sep)[-2].split("_")
    labels.append(label)
```

```python
labels
```

```python
data = np.array(data, dtype='float') / 255.0
labels = np.array(labels)
```

```python
labels
```

```python
mlb = MultiLabelBinarizer()
labels = mlb.fit_transform(labels)
```

```python
labels
```

```python
for (i, label) in enumerate(mlb.classes_):
    print("{}. {}".format(i + 1, label))
```

```python
(X_train, X_test, y_train, y_test) = train_test_split(data, labels, test_size = 0.2, random_state = 2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
gen = ImageDataGenerator(rotation_range=25, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.2, zoom_range=0.2, horizontal_flip=True, fill_mode="nearest")
```

```python
model = SmallerVGGNet.build(width=IMAGE_DIMS[1], height=IMAGE_DIMS[0], depth=IMAGE_DIMS[2], classes=len(mlb.classes_), finalAct="sigmoid")
```

```python
opt = Adam(learning_rate=LEARNING_RATE, decay=LEARNING_RATE / EPOCHS)
```

```python
model.compile(optimizer=opt, loss="binary_crossentropy", metrics=["accuracy"])
```

```python
gen.flow?
```

```python
trained = model.fit_generator(
    gen.flow(X_train, y_train, batch_size=BATCH_SIZE), 
    shuffle=True, validation_data=(X_test, y_test), 
    steps_per_epoch=len(X_train) // BATCH_SIZE, 
    epochs=EPOCHS, verbose=1)
```

```python
model.save('mlc_model')
```

```python
with open('labelbin', 'wb') as l:
    l.write(pickle.dumps(mlb))
```

```python
model.save_weights("mlc_model.h5")
```

```python
for key in trained.history:
    print(key)
```

```python
def plot():
    plt.style.use("ggplot")
    plt.figure()
    plt.plot(np.arange(0, EPOCHS), trained.history["loss"], label="Train Loss")
    plt.plot(np.arange(0, EPOCHS), trained.history["val_loss"], label="Test Loss")
    plt.plot(np.arange(0, EPOCHS), trained.history["accuracy"], label="Train Accuracy")
    plt.plot(np.arange(0, EPOCHS), trained.history["val_accuracy"], label="Test Accuracy")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss / Accuracy")
    plt.legend()
    plt.savefig("plot.png")
    plt.show()
```

```python
plot()
```
